```python
# 도로명 주소 -> 지번 주소
def makeUrl(search):
  url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=' + search
  return url

def crawlingAdderss(url) :
  html = requests.get(url)
  bs = BeautifulSoup(html.content, "html.parser")
  content = bs.find("span", {"class" :"_3rnws"})
  address = str(content)

  start = address.find('</span>')
  address = address[start+7:start+12]
  return address

def ChangeAddress(df, value) :
  df['우편주소'] = 0
  for i in range(len(df[value])) :
      search = df[value].iloc[i]
      url = makeUrl(search)
      address = crawlingAdderss(url)

      df['우편주소'].iloc[i] = address

ChangeAddress(cctv,'안심 주소')
cctv
```

```python
ChangeAddress(park,'공원주소')
park
```

```python
cctv['공원명'] = " "
cctv['고양이 급식소 설치 적합 여부'] = "비적합"

for i in range(len(cctv['우편주소'])) :
  for j in range(len(park['우편주소'])) :
    
    if(cctv['우편주소'].iloc[i] == park['우편주소'].iloc[j]) :
      cctv['고양이 급식소 설치 적합 여부'].iloc[i] = "적합"
      cctv['공원명'].iloc[i] = park['공원명'].iloc[j]

cctv
```

```python
result = cctv[cctv['고양이 급식소 설치 적합 여부'] == "적합"]
result['위도'] = result['위도'].astype(float)
result['경도'] = result['경도'].astype(float)
count = result['공원명'].value_counts()
count
```

```python
map = folium.Map(location = [37.55088815173826, 126.84951211051083], zoom_start = 13)
```

```python
for a in result.index :
  latitude = result.loc[a, "위도"]
  longtitude = result.loc[a,"경도"]
  folium.Marker([latitude, longtitude],
                popup = result.loc[a, "공원명"],
                tooltip = result.loc[a, "공원명"]).add_to(map)   
map
```
